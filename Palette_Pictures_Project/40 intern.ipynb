```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set default style
sns.set(style="whitegrid")

# Load the data
df = pd.read_csv("40 intern.csv")
```

```python
# Gender distribution
def plot_gender_distribution(df):
    gender_counts = df['gender(M/F)'].value_counts()
    gender_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, figsize=(6,6), title='Gender Distribution')
    plt.ylabel('')
    plt.tight_layout()
    plt.show()
```

```python
# Age distribution
def plot_age_distribution(df):
    age_counts = df['age(20,30,40,50,60)'].value_counts().sort_index()
    plt.figure(figsize=(8, 5))
    ax = sns.barplot(x=age_counts.index.astype(str), y=age_counts.values)
    ax.bar_label(ax.containers[0])
    plt.title('Age Distribution')
    plt.xlabel('Age Group')
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()
```

```python
# Role distribution
def plot_role_distribution(df):
    role_counts = df['Role'].value_counts()
    plt.figure(figsize=(10, 5))
    ax = sns.barplot(x=role_counts.index, y=role_counts.values)
    ax.bar_label(ax.containers[0])
    plt.title('Standard Role Distribution')
    plt.xlabel('Role')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
def plot_job_title_wordcloud(df):
    from wordcloud import WordCloud
    text = ' '.join(df['job title'].dropna().astype(str))
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('Job Title Word Cloud')
    plt.tight_layout()
    plt.show()

def print_top_10_job_words(df):
    from collections import Counter
    from wordcloud import STOPWORDS

    text = ' '.join(df['job title'].dropna().astype(str))
    words = text.lower().split()
    filtered_words = [w for w in words if w not in STOPWORDS]
    top_words = Counter(filtered_words).most_common(10)
    print("Top 10 most common words in job titles:")
    for word, count in top_words:
        print(f"{word}: {count}")
```

```python
# Stanford / Berkeley / NA distribution
def plot_school_background(df):
    school_counts = df['Stanford_Berkeley'].value_counts()
    plt.figure(figsize=(6, 5))
    ax = sns.barplot(x=school_counts.index, y=school_counts.values)
    ax.bar_label(ax.containers[0])
    plt.title('School Background Distribution')
    plt.xlabel('School')
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()
```

```python
plot_gender_distribution(df)
plot_age_distribution(df)
plot_role_distribution(df)
plot_job_title_wordcloud(df)
print_top_10_job_words(df)
plot_school_background(df)
```

```python
# Distribution of number of events organized
def plot_num_hosted_distribution(df):
    plt.figure(figsize=(8, 5))
    ax = sns.countplot(x="Num_Hosted", data=df)
    ax.bar_label(ax.containers[0])
    plt.title("Distribution of Number of Events Hosted")
    plt.xlabel("Number of Hosted Events")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

# Distribution of number of events attended
def plot_num_attended_distribution(df):
    plt.figure(figsize=(8, 5))
    ax = sns.countplot(x="Num_Attended", data=df)
    ax.bar_label(ax.containers[0])
    plt.title("Distribution of Number of Events Attended")
    plt.xlabel("Number of Attended Events")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

# Number of events attended with me
def plot_num_together_distribution(df):
    plt.figure(figsize=(8, 5))
    ax = sns.countplot(x="Num_Together", data=df)
    ax.bar_label(ax.containers[0])
    plt.title("Distribution of Events Attended Together with Me")
    plt.xlabel("Number of Events Together")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

# Number of cineAI events participated
def plot_cineai_event_distribution(df):
    plt.figure(figsize=(8, 5))
    ax = sns.countplot(x="cineAI_event", data=df)
    ax.bar_label(ax.containers[0])
    plt.title("Distribution of cineAI Event Participation")
    plt.xlabel("Number of cineAI Events")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

# Attendance rate analysis: proportion of approved participants who actually attended
def plot_approval_vs_come(df):
    approved_df = df[df['approval_status'] == 'approved']
    total = len(approved_df)
    came = approved_df['come?'].sum()
    not_came = total - came
    plt.figure(figsize=(6, 5))
    plt.pie([came, not_came], labels=['Came', 'Did Not Come'], autopct='%1.1f%%', startangle=90)
    plt.title("Attendance Rate among Approved Users")
    plt.tight_layout()
    plt.show()

plot_num_hosted_distribution(df)
plot_num_attended_distribution(df)
plot_num_together_distribution(df)
plot_cineai_event_distribution(df)
plot_approval_vs_come(df)
```

```python
# Ticket type distribution
def plot_ticket_type_distribution(df):
    ticket_counts = df['ticket_type'].value_counts()
    plt.figure(figsize=(8, 5))
    ax = sns.barplot(x=ticket_counts.index, y=ticket_counts.values)
    ax.bar_label(ax.containers[0])
    plt.title("Distribution of Ticket Types")
    plt.xlabel("Ticket Type")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Participation priority distribution
def plot_participation_priority(df):
    plt.figure(figsize=(6, 5))
    ax = sns.countplot(x='Participation Priority', data=df, order=sorted(df['Participation Priority'].dropna().unique()))
    ax.bar_label(ax.containers[0])
    plt.title("Participation Priority Distribution")
    plt.xlabel("Priority Level")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

# Willingness to pay distribution
def plot_willingness_to_pay(df):
    plt.figure(figsize=(6, 5))
    ax = sns.countplot(x='Willingness to pay', data=df, order=sorted(df['Willingness to pay'].dropna().unique()))
    ax.bar_label(ax.containers[0])
    plt.title("Willingness to Pay Distribution")
    plt.xlabel("Willingness Level")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

plot_ticket_type_distribution(df)
plot_participation_priority(df)
plot_willingness_to_pay(df)
```

```python
def analyze_paid_amount(df):
    # Filter paid users
    df['amount_clean'] = df['amount'].replace('[\$,]', '', regex=True).astype(float)
    paid_df = df[df['amount_clean'] > 0]

    mean_amt = paid_df['amount_clean'].mean()
    median_amt = paid_df['amount_clean'].median()
    mode_amt = paid_df['amount_clean'].mode().iloc[0] if not paid_df['amount_clean'].mode().empty else None

    print(f"平均付款金额: ${mean_amt:.2f}")
    print(f"中位数付款金额: ${median_amt:.2f}")
    print(f"众数付款金额: ${mode_amt:.2f}")

    plt.figure(figsize=(8, 5))
    sns.histplot(paid_df['amount_clean'], bins=10, kde=True)
    plt.title("Distribution of Paid Ticket Amounts")
    plt.xlabel("Amount Paid ($)")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()
```

```python
analyze_paid_amount(df)
```

```python
# Whether the user is a CineAI member
def plot_membership_distribution(df):
    # Standardize yes/no/blank values to Member / Non-member
    df['member_clean'] = df['member of the CineAI Founders Hub?'].astype(str).str.lower().str.strip()
    df['member_clean'] = df['member_clean'].replace({'': 'no', 'nan': 'no'})
    df['member_str'] = df['member_clean'].map({'yes': 'Member', 'no': 'Non-member'})

    member_counts = df['member_str'].value_counts()

    plt.figure(figsize=(6, 5))
    ax = sns.barplot(x=member_counts.index, y=member_counts.values)
    ax.bar_label(ax.containers[0])
    plt.title("CineAI Membership Distribution")
    plt.xlabel("Membership")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(6, 5))
    plt.pie(member_counts, labels=member_counts.index, autopct='%1.1f%%', startangle=90)
    plt.title("CineAI Membership Distribution (Pie)")
    plt.tight_layout()
    plt.show()

# Industry interest distribution
def plot_industry_interest(df):
    # Find column names containing "industry"
    column_name = [col for col in df.columns if 'industry' in col.lower()][0]
    industries = df[column_name].dropna().astype(str).str.split(',')
    all_industries = [i.strip() for sublist in industries for i in sublist]
    industry_counts = pd.Series(all_industries).value_counts().head(10)

    plt.figure(figsize=(8, 5))
    ax = sns.barplot(x=industry_counts.values, y=industry_counts.index)
    ax.bar_label(ax.containers[0])
    plt.title("Top Interested Industries")
    plt.xlabel("Count")
    plt.ylabel("Industry")
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(6, 5))
    plt.pie(industry_counts, labels=industry_counts.index, autopct='%1.1f%%', startangle=90)
    plt.title("Top Interested Industries (Pie)")
    plt.tight_layout()
    plt.show()

# Whether they are using or exploring AI
def plot_ai_usage(df):
    usage = df['currently using or exploring AI for content creation or brand building?'].dropna().str.strip()
    usage_counts = usage.value_counts()

    plt.figure(figsize=(8, 5))
    ax = sns.barplot(x=usage_counts.index, y=usage_counts.values)
    ax.bar_label(ax.containers[0])
    plt.title("AI Usage for Content/Branding")
    plt.xlabel("Usage Type")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(6, 5))
    plt.pie(usage_counts, labels=usage_counts.index, autopct='%1.1f%%', startangle=90)
    plt.title("AI Usage for Content/Branding (Pie)")
    plt.tight_layout()
    plt.show()

# Topic interest distribution
def plot_topic_interest(df):
    topics = df['topics most interested in'].dropna().str.split(',')
    all_topics = [t.strip() for sublist in topics for t in sublist]
    topic_counts = pd.Series(all_topics).value_counts().head(10)

    plt.figure(figsize=(8, 5))
    ax = sns.barplot(x=topic_counts.values, y=topic_counts.index)
    ax.bar_label(ax.containers[0])
    plt.title("Top Interested Topics")
    plt.xlabel("Count")
    plt.ylabel("Topic")
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(6, 5))
    plt.pie(topic_counts, labels=topic_counts.index, autopct='%1.1f%%', startangle=90)
    plt.title("Top Interested Topics (Pie)")
    plt.tight_layout()
    plt.show()

# Follow-up intent distribution
def plot_followup(df):
    follow = df['followup allowed?'].dropna().str.strip()
    follow_counts = follow.value_counts()

    plt.figure(figsize=(6, 5))
    ax = sns.barplot(x=follow_counts.index, y=follow_counts.values)
    ax.bar_label(ax.containers[0])
    plt.title("Follow-up Permission")
    plt.xlabel("Follow-up Allowed")
    plt.ylabel("Count")
    plt.xticks(rotation=30, ha='right')
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(6, 5))
    plt.pie(follow_counts, labels=follow_counts.index, autopct='%1.1f%%', startangle=90)
    plt.title("Follow-up Permission (Pie)")
    plt.tight_layout()
    plt.show()

plot_membership_distribution(df)
plot_industry_interest(df)
plot_ai_usage(df)
plot_topic_interest(df)
plot_followup(df)
```

```python
# Response status: Approve vs Decline
def plot_response_distribution(df):
    response = df['approval_status'].dropna().str.strip().str.lower()
    response_counts = response.value_counts()

    plt.figure(figsize=(6, 5))
    ax = sns.barplot(x=response_counts.index, y=response_counts.values)
    ax.bar_label(ax.containers[0])
    plt.title("Invitation Response Distribution")
    plt.xlabel("Response Type")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

plot_response_distribution(df)
```

```python
# Payment status
def plot_payment_status(df):
    df['amount_clean'] = df['amount'].replace('[\$,]', '', regex=True)
    df['amount_clean'] = pd.to_numeric(df['amount_clean'], errors='coerce')
    paid_status = df['amount_clean'] > 0

    paid_counts = paid_status.value_counts().rename({True: 'Paid', False: 'Free'})

    plt.figure(figsize=(6, 5))
    ax = sns.barplot(x=paid_counts.index, y=paid_counts.values)
    ax.bar_label(ax.containers[0])
    plt.title("Ticket Payment Status")
    plt.xlabel("Status")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(6, 5))
    plt.pie(paid_counts.values, labels=paid_counts.index, autopct='%1.1f%%', startangle=90)
    plt.title("Percentage of Users Who Paid for Tickets")
    plt.tight_layout()
    plt.show()
```

```python
plot_payment_status(df)
```
