```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Upload local Excel file
uploaded = files.upload()
fname = list(uploaded.keys())[0]

# Read Excel file
df = pd.read_excel("Inequality Trend.xlsx")
df.columns = df.columns.str.strip()
df.dropna(how='all', inplace=True)

# Plot
plt.figure(figsize=(10,5))
if "GINI Index" in df.columns:
    plt.plot(df["Years"], df["GINI Index"], marker='o', label="GINI Index")
if "THEIL Index" in df.columns:
    plt.plot(df["Years"], df["THEIL Index"], marker='o', label="THEIL Index")

plt.title("GINI Index vs THEIL Index Over Time")
plt.xlabel("Year")
plt.ylabel("Index Value")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
