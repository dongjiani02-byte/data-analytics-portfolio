```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the data
file_path = 'Sephora_Age_Analysis.csv'
df = pd.read_csv(file_path)

# Stacked Bar Chart
age_columns = ['Age_35to44', 'Age_18to24', 'Age_25to34', 'Age_45to54', 'Age_over54', 'Age_13to17']
df_age_distribution = df.groupby('Name')[age_columns].sum()

df_age_distribution.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title("Age Group Distribution for Products (Stacked Bar Chart)")
plt.xlabel("Product Name")
plt.ylabel("Customer Count")
plt.legend(title="Age Groups", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df_age_distribution, annot=False, cmap="YlGnBu")
plt.title("Relationship Between Products and Age Groups (Heatmap)")
plt.xlabel("Age Groups")
plt.ylabel("Product Name")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the data
file_path = 'Sephora_Age_Analysis.csv'
df = pd.read_csv(file_path)

# Select numerical columns (age group data)
age_columns = ['Age_35to44', 'Age_18to24', 'Age_25to34', 'Age_45to54', 'Age_over54', 'Age_13to17']
df_age_data = df[age_columns]

# Calculate the correlation matrix
correlation_matrix = df_age_data.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Age Groups")
plt.tight_layout()
plt.show()
```

```python
# Adjusting the heatmap size and layout for better display
plt.figure(figsize=(8, 6))  # Smaller figure size
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", cbar_kws={'shrink': 0.8})
plt.title("Correlation Heatmap of Age Groups")
plt.tight_layout()  # Automatically adjusts subplot parameters to fit within the figure
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = 'Sephora_Category_Analysis.csv'
df = pd.read_csv(file_path)

# Count products by category_1, category_2, and category_3
category_counts = df[['category_1', 'category_2', 'category_3']].melt().value_counts().reset_index()
category_counts.columns = ['Category_Level', 'Category_Name', 'Count']

# Plot a stacked bar chart
category_grouped = category_counts.pivot(index='Category_Name', columns='Category_Level', values='Count')

category_grouped.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title("Distribution of Products Across Categories")
plt.xlabel("Categories")
plt.ylabel("Product Count")
plt.xticks(rotation=90)
plt.legend(title="Category Level")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_path = 'Sephora_Category_Analysis.csv'
df = pd.read_csv(file_path)
category_data = df.groupby('Name')[['category_1', 'category_2', 'category_3']].apply(
    lambda x: x.notnull().sum()
)
plt.figure(figsize=(12, 8))
bar_width = 0.8

for i, category in enumerate(['category_1', 'category_2', 'category_3']):
    plt.bar(
        category_data.index,
        category_data[category],
        bottom=category_data.iloc[:, :i].sum(axis=1) if i > 0 else None,
        label=category
    )

plt.xlabel('Product Name', fontsize=10)
plt.ylabel('Category Count', fontsize=10)
plt.title('Relationship Between Products and Categories', fontsize=12)
plt.xticks(rotation=90, ha='center', fontsize=8)
plt.legend(title="Category Levels")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'Sephora_Category_Analysis.csv'
df = pd.read_csv(file_path)
def plot_category_bar_chart(df, category_column):
    category_counts = df[category_column].value_counts()
    plt.figure(figsize=(12, 6))
    plt.bar(category_counts.index, category_counts.values, color='lightblue')
    plt.xlabel(f'{category_column} Names', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.title(f'Count of Each Name in {category_column}', fontsize=14)
    plt.xticks(rotation=45, ha='right', fontsize=10)
    plt.tight_layout()
    plt.show()
plot_category_bar_chart(df, 'category_1')
plot_category_bar_chart(df, 'category_2')
plot_category_bar_chart(df, 'category_3')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'Sephora_Category_Love_Reviews_Analysis.csv'
df = pd.read_csv(file_path)
categories = pd.concat([
    df[['category_1', 'love_count', 'reviews_count']].rename(columns={'category_1': 'category'}),
    df[['category_2', 'love_count', 'reviews_count']].rename(columns={'category_2': 'category'}),
    df[['category_3', 'love_count', 'reviews_count']].rename(columns={'category_3': 'category'}),
])
category_summary = categories.groupby('category')[['love_count', 'reviews_count']].sum()

plt.figure(figsize=(12, 6))
plt.bar(category_summary.index, category_summary['love_count'], color='skyblue')
plt.xlabel('Category', fontsize=12)
plt.ylabel('Total Love Count', fontsize=12)
plt.title('Total Love Count by Category', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
plt.bar(category_summary.index, category_summary['reviews_count'], color='lightgreen')
plt.xlabel('Category', fontsize=12)
plt.ylabel('Total Reviews Count', fontsize=12)
plt.title('Total Reviews Count by Category', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the data
file_path = 'Sephora - Sephora_Price_size_Analysis.csv'
df = pd.read_csv(file_path)

# Scatter Plot: Price vs size_oz
plt.figure(figsize=(10, 6))
plt.scatter(df['size_oz'], df['Price'], alpha=0.7, color='skyblue')
plt.title('Price vs Size (oz)', fontsize=14)
plt.xlabel('Size (oz)', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.grid(alpha=0.3)

# Add Regression Line
m, b = np.polyfit(df['size_oz'], df['Price'], 1)  # Linear fit (y = mx + b)
plt.plot(df['size_oz'], m * df['size_oz'] + b, color='red', label='Regression Line')
plt.legend()
plt.tight_layout()
plt.show()

# Correlation: Price vs size_oz
correlation_price_size_oz = df['Price'].corr(df['size_oz'])
print(f"Correlation between Price and Size (oz): {correlation_price_size_oz:.2f}")

# Scatter Plot: Price vs size_mL
plt.figure(figsize=(10, 6))
plt.scatter(df['size_mL'], df['Price'], alpha=0.7, color='lightgreen')
plt.title('Price vs Size (mL)', fontsize=14)
plt.xlabel('Size (mL)', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.grid(alpha=0.3)

# Add Regression Line
m, b = np.polyfit(df['size_mL'], df['Price'], 1)  # Linear fit (y = mx + b)
plt.plot(df['size_mL'], m * df['size_mL'] + b, color='red', label='Regression Line')
plt.legend()
plt.tight_layout()
plt.show()

# Correlation: Price vs size_mL
correlation_price_size_ml = df['Price'].corr(df['size_mL'])
print(f"Correlation between Price and Size (mL): {correlation_price_size_ml:.2f}")
```
